* **What's Singleton pattern? How to implement it?**  
    - Ensures only one instance of a class exists.
    - How to implement:
	    - Store the single instance in a class variable.
	    - When creating a new object, check if one already exists.
	    - If yes, return the same one instead of creating a new one.

```python
class Singleton:
    _instance = None
    
    def __new__(cls):
        if cls._instance is None:          # if not created yet
            cls._instance = super().__new__(cls)
        return cls._instance

# a = Singleton() and b = Singleton() will point to the same object.
a = Singleton()
b = Singleton()
a is b
```

* **What's Factory pattern? How to implement it?**
    - create objects without showing the creation logic to the user.
    - How to implement:
	    - Define a factory class or method.
	    - Inside it, decide which class to create based on some input (like a type or name).
	    - Return that object.

```python
class Dog: pass
class Cat: pass

class AnimalFactory:
    @staticmethod
    def create_animal(type):
        if type == "dog":
            return Dog()
        elif type == "cat":
            return Cat()

# Calling AnimalFactory.create_animal("dog") gives you a Dog object, 
# but you dont need to know how its made.
dog_1 = AnimalFactory.create_animal("dog")
```

* **What are abstract base classes (ABC) in Python?**
    - define a template (interface) for other classes.
    - How to implement:

	    - Create a base class that inherits from ABC.
	    - Mark methods that must be overridden with @abstractmethod.
