## Concept Questions

### **Difference between authentication and authorization**
- authentication verify who you are, check the identity  
- authorization verify permissions, what you can access

### **What are HTTP cookies? How do they work and what are their main use cases?**
what are cookies:   
small pieces of data stored in the browser and auto sent to the server with every request to the same domain   

how they work:  
1.	Server sends a Set-Cookie header
2.	Browser creates and stores the cookie
3.	Browser automatically includes it in future requests

main use cases:   
- Session management (logins)
- Remembering preferences (theme, language)
- Tracking (analytics)

### **What are the limitations of cookies**
- Size limit: ~4 KB per cookie
- Cannot store complex objects (must be stringified)
- Count limit: ~50 per site
- Sent automatically → can increase request size
- Security risks if not using flags:
  - HttpOnly
  - Secure
  - SameSite
- Subject to browser restrictions (privacy modes, blocking)

### **What is JWT and how does it work?**
JSON Web Token is a signed token that contains user data (claims)  

How it works  
1. User logs in
2. Server creates a token (signed with secret/private key)
3. Client stores the token (usually in localStorage or cookies)
4. Client sends requests with token in headers:  Authorization: Bearer <token>
5. Server decode token and verifies signature — no DB lookup needed
6. If valid, user is authenticated

### **What are the advantages and disadvantages of using JWT compared to traditional session-based authentication?**
Advantages
- Stateless (server doesn’t store sessions)
- Faster — no DB lookup for every request
- Scales easily across microservices
- Works across domains / APIs

✖ Disadvantages
- Cannot be easily invalidated until expiration
- Larger token size (sent every request) than session id
- More vulnerable if stored in localStorage
- Rotation/blacklist logic is complex

### **How do you invalidate or blacklist JWT tokens?**
Ways to invalidate:
1.	Short expiration time (access token)
2.	Maintain a blacklist/denylist (store jti or token hash)
3.	Change server secret key (invalidates all tokens)
4.	Use refresh tokens + rotation
5.	Store “token version” in DB and increment it on logout

Best practice = two tokens pattern: short-lived access token + refresh token rotation.

### **What is password hashing and why is it important?**
encoding a plain password into an irreversible string using algorithms like:
- bcrypt
- Argon2 (best)
- PBKDF2
- scrypt

Why important?
- If database leaks, raw passwords are not exposed
- Hashes cannot be reversed, so more secure
> - Brute-forcing is slowed with salting + work factor

### **What is the access / refresh token pattern**
A secure authentication pattern using two tokens:    

1. Access Token
	- Short-lived (5–15 mins)
	- Sent with every API request
	- If expired, use refresh token to get a new one
	- If stolen → limited damage

2. Refresh Token
	- Long-lived (days/weeks)
	- Stored in secure cookie
	- Used to generate new access tokens
	- If expired, need to login again
	- Never exposed to JavaScript (HttpOnly)

### **What is the OAuth2 flow**
A protocol that allows users to grant third-party applications **limited access** to their resources on another service, **without sharing passwords**. like sign in with Google:  
1.	User clicks Login with Google/GitHub
2.	Redirect to OAuth provider’s login page
3.	User logs in and approves the app
4.	Provider redirects back with an authorization code
5.	Backend exchanges the code for an access token
6.	Backend uses token to access user data from provider
7.	Backend creates a session/JWT for your app
