# Quiz 1:  
Merge Intervals - Coding Challenge  
Problem Description  
Given an array of intervals where intervals[i] = [start_i, end_i], merge all overlapping intervals and return an array of non-overlapping intervals.  

Examples  
Example 1:   
Input: [[1,3], [2,6], [8,10], [15,18]]   
Output: [[1,6], [8,10], [15,18]]   
Example 2:   
Input: [[1,4], [4,5]]   
Output: [[1,5]]  
Example 3:   
Input: [[1,4], [0,2], [3,5]]   
Output: [[0,5]]  

Constraints   
●	1 <= intervals.length <= 10^4
●	0 <= start_i <= end_i <= 10^4

Test Cases   
from typing import List   

assert merge([[1,3],[2,6],[8,10],[15,18]]) == [[1,6],[8,10],[15,18]]   
assert merge([[1,4],[4,5]]) == [[1,5]]    
assert merge([[1,4],[0,4]]) == [[0,4]]   
Expected: O(n log n) time, O(n) space   

```python
from typing import List

def merge(intervals: List[List[int]]) -> List[List[int]]:
    """
    Merges all overlapping intervals.
    
    Args:
        intervals: List of intervals [start, end]
    Returns:
        List of merged non-overlapping intervals
    """
    intervals = sorted(intervals)
    n = len(intervals)

    ans = []
    ans.append(intervals[0])
    i, j = 0, 1
    while j < n:
        if intervals[j][0] <= ans[i][1]:
            ans[i] = [ans[i][0], intervals[j][1]]
        else:
            ans.append(intervals[j])
            i += 1
        
        j += 1
    
    return ans

print(merge([[1,3],[2,6],[8,10],[15,18]]) == [[1,6],[8,10],[15,18]])
print(merge([[1,4],[4,5]]) == [[1,5]])
print(merge([[1,4],[0,4]]) == [[0,4]])
```
